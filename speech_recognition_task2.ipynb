```python
# Step 1: Libraries install
!pip install SpeechRecognition pydub
!apt-get install -y ffmpeg

# Step 2: Libraries import
import speech_recognition as sr
from pydub import AudioSegment
import os

# Step 3: Check if user uploaded audio file
audio_filename = "user_audio.wav"  # change if user uploaded a file

if not os.path.exists(audio_filename):
    # Download sample audio if no file uploaded
    print("No audio uploaded, downloading sample audio...")
    !wget -O sample_audio.wav https://www2.cs.uic.edu/~i101/SoundFiles/CantinaBand60.wav
    audio_filename = "sample_audio.wav"

# Step 4: Convert MP3 to WAV if needed
if audio_filename.endswith(".mp3"):
    sound = AudioSegment.from_file(audio_filename)
    audio_filename = "converted_audio.wav"
    sound.export(audio_filename, format="wav")
    print(f"Converted audio to WAV: {audio_filename}")

# Step 5: Initialize recognizer
r = sr.Recognizer()
audio_file = sr.AudioFile(audio_filename)

# Step 6: Read audio file
with audio_file as source:
    audio_data = r.record(source)  # read entire audio

# Step 7: Recognize speech using Google API
try:
    text = r.recognize_google(audio_data)
    print("✅ Recognized Text:\n", text)
except sr.UnknownValueError:
    print("⚠️ Could not understand audio")
except sr.RequestError as e:
    print("⚠️ Could not request results; check internet connection")
```

```python
import speech_recognition as sr
from pydub import AudioSegment
```

```python
# Load .ogg
sound = AudioSegment.from_file("kcc.ogg")

# Export to .wav
sound.export("kcc.wav", format="wav")
print("✅ Converted .ogg to .wav")
```

```python
r = sr.Recognizer()
with sr.AudioFile("kcc.wav") as source:
    audio_data = r.record(source)

try:
    text = r.recognize_google(audio_data)
    print("✅ Recognized Text:\n", text)
except sr.UnknownValueError:
    print("⚠️ Could not understand audio")
except sr.RequestError as e:
    print("⚠️ Could not request results; check internet connection")
```
